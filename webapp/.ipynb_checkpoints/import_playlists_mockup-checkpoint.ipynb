```python
import datetime
import os
from stat import *
from shutil import copyfile
from config import Config
import pandas
import xlrd
import fnmatch
```

```python
base = datetime.datetime.today()
date_list = [(base + datetime.timedelta(days=x)) for x in range(5)]

media_files=[]
try:
    for root, dirnames, filenames in os.walk(Config.MEDIA_FOLDER):
        for filename in fnmatch.filter(filenames, "*.mxf"):
            media_files.append(os.path.join(root, filename))
    print("Found {} mxf files in {}".format(len(media_files),Config.MEDIA_FOLDER))
except Exception as e:
    print('Failed to check missing media files: '+ str(e))
    raise e
    

for lf in Config.LOCALIZED_FEEDS:
    for d in date_list:
        src_xls_filename=os.path.join(lf["xls"],d.strftime("%Y"),'Sent',d.strftime("%m")+"_"+d.strftime('%B'),d.strftime("%Y-%m-%d")+"_"+lf["log"]+".xls")
        dst_xls_filename=os.path.join(Config.STREAMS_ROOT,lf["dir"],"commercials","playlists",d.strftime("%Y-%m-%d"),d.strftime("%Y-%m-%d")+"_"+lf["log"]+".xls")
        print(src_xls_filename)
        if os.path.isfile(src_xls_filename):
            src_stat=os.stat(src_xls_filename)
            if os.path.isfile(dst_xls_filename):
                dst_stat=os.stat(dst_xls_filename)
                if src_stat[ST_SIZE]==dst_stat[ST_SIZE] and src_stat[ST_MTIME]<dst_stat[ST_MTIME]:
                    #Likely that the two files are the same, skip to copy it....
                    continue          
            os.makedirs(os.path.dirname(dst_xls_filename), exist_ok=True)
            copyfile(src_xls_filename,dst_xls_filename)
            print("{} copied to {}".format(src_xls_filename,dst_xls_filename))
            missing=set()
            try:
                df=pandas.read_excel(dst_xls_filename)
                for c in df["Cart Number"].unique():
                    found_cart=False
                    for m in media_files:
                        if c in m:
                            # print("{} found in {}".format(c,m))
                            found_cart=True
                            break
                    if not found_cart:
                        # print("Adding {} to missing list.".format(c))
                        missing.add(c)
            except Exception as e:
                print('Failed to check missing media files: '+ str(e))
                print("Deleting {}".format(dst_xls_filename))
                os.remove(dst_xls_filename)
                continue
            with open(os.path.join(os.path.dirname(dst_xls_filename),".missing_files"), 'w') as f:
                for item in missing:
                    f.write("%s\n" % item)
            f.close()
```
